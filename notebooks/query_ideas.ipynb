```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('../db/footbase_big5.db')

# Check xG coverage by league
print("xG Coverage by League:")
coverage = pd.read_sql_query("""
    SELECT league, 
           COUNT(*) as total_matches,
           SUM(CASE WHEN home_xg IS NOT NULL THEN 1 ELSE 0 END) as with_xg,
           ROUND(100.0 * SUM(CASE WHEN home_xg IS NOT NULL THEN 1 ELSE 0 END) / COUNT(*), 1) as percentage,
           ROUND(AVG(home_xg), 2) as avg_home_xg,
           ROUND(AVG(away_xg), 2) as avg_away_xg
    FROM matches 
    GROUP BY league
    ORDER BY league
""", conn)
print(coverage.to_string(index=False))

# Sample matches with xG from each league
print("\n\nSample matches with xG from each league:")
sample = pd.read_sql_query("""
    SELECT league, date, home_team, away_team, 
           home_goals, away_goals, home_xg, away_xg
    FROM matches 
    WHERE home_xg IS NOT NULL
    GROUP BY league
    LIMIT 5
""", conn)
print(sample.to_string(index=False))

conn.close()
```
